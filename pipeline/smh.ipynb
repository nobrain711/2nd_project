```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/BankChurners.csv')
df
```

```python
df = df.drop([
    'CLIENTNUM',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'
], axis=1)
df.head()
```

```python
rename_dict = {
    "Attrition_Flag":"churn",
    "Customer_Age":"age",
    "Gender":"gender",
    "Dependent_count":"dependents",
    "Education_Level":"education",
    "Marital_Status":"marital",
    "Income_Category":"income",
    "Card_Category":"card_type",
    "Months_on_book":"relationship_months",
    "Total_Relationship_Count":"product_count",
    "Months_Inactive_12_mon":"inactive_months",
    "Contacts_Count_12_mon":"contact_count",
    "Credit_Limit":"credit_limit",
    "Total_Revolving_Bal":"revolving_balance",
    "Avg_Open_To_Buy":"available_credit",
    "Total_Amt_Chng_Q4_Q1":"amount_change",
    "Total_Ct_Chng_Q4_Q1":"count_change",
    "Total_Trans_Amt":"transaction_amount",
    "Total_Trans_Ct":"transaction_count",
    "Avg_Utilization_Ratio":"utilization_ratio"
}

df = df.rename(columns=rename_dict)
df.head()
```

```python
gender_ser = df['gender'].copy()
gender_ser = gender_ser.map(lambda x: 0 if x == 'M' else 1)
# gender_ser

gender_df = pd.DataFrame({
    'gender':df['gender'],
    'gender_encoded':gender_ser
})
gender_df
```

```python
"""
| income_id | income_label   | min    | max    |
| --------- | -------------- | ------ | ------ |
| 1         | Less than $40K | 0      | 40000  |
| 2         | $40K-$60K      | 40000  | 60000  |
| 3         | $60K-$80K      | 60000  | 80000  |
| 4         | $80K-$120K     | 80000  | 120000 |
| 5         | $120K+         | 120000 | NULL   |
| 6         | Unknown        | NULL   | NULL   |
"""
```

```python
def income_match(x):
    match x:
        case "Less than $40K":
            return 1
        case "$40K - $60K":
            return 2
        case "$60K - $80K":
            return 3
        case "$80K - $120K":
            return 4
        case "$120K +":
            return 5
        case "Unknown":
            return 6
```

```python
income_ser = df['income'].copy()
income_ser = income_ser.apply(income_match)

income_df = pd.DataFrame({
    'income':df['income'],
    'income_encoded':income_ser
})
income_df
```

```python
"""
| card_type_id | card_type_label |
| ------------ | --------------- |
| 1            | Blue            |
| 2            | Silver          |
| 3            | Gold            |
| 4            | Platinum        |
"""
```

```python
def card_type_match(x):
    match x:
        case "Blue":
            return 1
        case "Silver":
            return 2
        case "Gold":
            return 3
        case "Platinum":
            return 4
```

```python
card_ser = df['card_type'].copy()
card_ser = card_ser.apply(card_type_match)

card_df = pd.DataFrame({
    'card_type':df['card_type'],
    'card_type_encoded':card_ser
})
card_df
```

```python
"""
| marital_id | marital_label |
| ---------- | ------------- |
| 1          | Single        |
| 2          | Married       |
| 3          | Divorced      |
| 4          | Unknown       |
"""
```

```python
def marital_match(x):
    match x:
        case "Single":
            return 1
        case "Married":
            return 2
        case "Divorced":
            return 3
        case "Unknown":
            return 4
```

```python
marital_ser = df['marital'].copy()
marital_ser = marital_ser.apply(marital_match)

marital_df = pd.DataFrame({
    'marital':df['marital'],
    'marital_encoded':marital_ser
})
marital_df
```

```python
"""
| education_id | education_label |
| ------------ | --------------- |
| 1            | Uneducated      |
| 2            | High School     |
| 3            | College         |
| 4            | Graduate        |
| 5            | Post-Graduate   |
| 6            | Doctorate       |
| 7            | Unknown         |
"""
```

```python
def education_match(x):
    match x:
        case "Uneducated":
            return 1
        case "High School":
            return 2
        case "College":
            return 3
        case "Graduate":
            return 4
        case "Post-Graduate":
            return 5
        case "Doctorate":
            return 6
        case "Unknown":
            return 7
```

```python
education_ser = df['education'].copy()
education_ser = education_ser.apply(education_match)

education_df = pd.DataFrame({
    'education':df['education'],
    'education_encoded':education_ser
})
education_df
```

```python
churn_ser = df['churn'].copy()
churn_ser = churn_ser.map(lambda x: 1 if x == 'Existing Customer' else 0)
# churn_ser

churn_df = pd.DataFrame({
    'churn':df['churn'],
    'churn_encoded':churn_ser
})
churn_df
```

```python
# =========================================================
# 카드사 고객 세그먼트 + churn 인사이트 분석
# 발표 구조:
# 1) 고객 활동 수준 정의
# 2) 4개 세그먼트 도출
# 3) 세그먼트별 churn 비교
# 4) 행동 변화(count_change) 해석
# 5) 미사용(inactive_months) 의미 재해석
# 6) 고객 중요도 연결
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

# ---------------------------------------------------------
# 1. 데이터 불러오기
# ---------------------------------------------------------
df = pd.read_csv("preprocessed_BankChurners.csv")

print("=" * 70)
print("[기본 정보]")
print(df.shape)
print(df.head())
print(df.columns.tolist())

# ---------------------------------------------------------
# 2. 발표에 사용할 핵심 변수 확인
# - 활동 수준 핵심 변수:
#   transaction_count, transaction_amount, credit_limit
# - 보조 변수:
#   utilization_ratio, inactive_months, count_change
# - 타깃:
#   churn
# ---------------------------------------------------------
use_cols = [
    "churn",
    "transaction_count",
    "transaction_amount",
    "credit_limit",
    "utilization_ratio",
    "inactive_months",
    "count_change",
    "contact_count",
    "available_credit",
    "revolving_balance"
]

data = df[use_cols].copy()

print("=" * 70)
print("[분석용 데이터 요약]")
print(data.describe().T)

# ---------------------------------------------------------
# 3. 기본 EDA
# ---------------------------------------------------------
print("=" * 70)
print("[전체 churn 비율]")
print(data["churn"].value_counts(normalize=True).sort_index())

fig, axes = plt.subplots(2, 3, figsize=(16, 9))
plot_cols = [
    "transaction_count",
    "transaction_amount",
    "credit_limit",
    "utilization_ratio",
    "inactive_months",
    "count_change"
]

for ax, col in zip(axes.flatten(), plot_cols):
    sns.histplot(data[col], kde=True, ax=ax)
    ax.set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 4. 고객 활동 수준(Activity Score) 정의
# - transaction_count, transaction_amount, credit_limit를 표준화
# - 평균을 Activity Score로 사용
# ---------------------------------------------------------
activity_cols = ["transaction_count", "transaction_amount", "credit_limit"]

scaler = StandardScaler()
scaled_activity = scaler.fit_transform(data[activity_cols])

scaled_activity_df = pd.DataFrame(
    scaled_activity,
    columns=[f"{c}_z" for c in activity_cols],
    index=data.index
)

data = pd.concat([data, scaled_activity_df], axis=1)
data["activity_score"] = scaled_activity_df.mean(axis=1)

print("=" * 70)
print("[Activity Score 요약]")
print(data["activity_score"].describe())

# ---------------------------------------------------------
# 5. 4개 세그먼트 도출
# - activity_score 기준 사분위수로 구분
# ---------------------------------------------------------
data["segment"] = pd.qcut(
    data["activity_score"],
    4,
    labels=["Dormant", "Casual", "Active", "Premium"]
)

print("=" * 70)
print("[세그먼트 분포]")
print(data["segment"].value_counts())

# ---------------------------------------------------------
# 6. 세그먼트별 핵심 지표 비교
# ---------------------------------------------------------
segment_summary = data.groupby("segment").agg({
    "transaction_count": ["mean", "median"],
    "transaction_amount": ["mean", "median"],
    "credit_limit": ["mean", "median"],
    "utilization_ratio": ["mean", "median"],
    "inactive_months": ["mean", "median"],
    "count_change": ["mean", "median"],
    "contact_count": ["mean", "median"],
    "churn": ["mean"]
}).round(3)

print("=" * 70)
print("[세그먼트별 요약표]")
print(segment_summary)

# 보기 편하게 평탄화
segment_table = data.groupby("segment").agg({
    "transaction_count": "mean",
    "transaction_amount": "mean",
    "credit_limit": "mean",
    "utilization_ratio": "mean",
    "inactive_months": "mean",
    "count_change": "mean",
    "contact_count": "mean",
    "churn": "mean"
}).round(3).reset_index()

print("=" * 70)
print("[발표용 세그먼트 평균표]")
print(segment_table)

# ---------------------------------------------------------
# 7. 세그먼트별 시각화
# ---------------------------------------------------------
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

sns.barplot(data=data, x="segment", y="transaction_count", ax=axes[0, 0])
axes[0, 0].set_title("Avg Transaction Count by Segment")

sns.barplot(data=data, x="segment", y="transaction_amount", ax=axes[0, 1])
axes[0, 1].set_title("Avg Transaction Amount by Segment")

sns.barplot(data=data, x="segment", y="credit_limit", ax=axes[0, 2])
axes[0, 2].set_title("Avg Credit Limit by Segment")

sns.barplot(data=data, x="segment", y="utilization_ratio", ax=axes[1, 0])
axes[1, 0].set_title("Avg Utilization Ratio by Segment")

sns.barplot(data=data, x="segment", y="inactive_months", ax=axes[1, 1])
axes[1, 1].set_title("Avg Inactive Months by Segment")

sns.barplot(data=data, x="segment", y="churn", ax=axes[1, 2])
axes[1, 2].set_title("Churn Rate by Segment")

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 8. 핵심 인사이트 1
# - 세그먼트별 churn 비율
# ---------------------------------------------------------
segment_churn = (
    data.groupby("segment")["churn"]
    .mean()
    .sort_values(ascending=False)
    .round(4)
)

print("=" * 70)
print("[세그먼트별 churn 비율]")
print(segment_churn)

# ---------------------------------------------------------
# 9. 핵심 인사이트 2
# - 같은 활동 감소(count_change)라도 세그먼트에 따라 의미가 다른가
# - 각 세그먼트 안에서 count_change를 3구간으로 나눠 churn 비교
# ---------------------------------------------------------
temp_list = []

for seg in data["segment"].unique():
    seg_df = data[data["segment"] == seg].copy()
    
    # count_change를 세 구간(낮음/중간/높음)으로 분할
    seg_df["count_change_bin"] = pd.qcut(
        seg_df["count_change"],
        3,
        labels=["Low Change", "Mid Change", "High Change"],
        duplicates="drop"
    )
    
    tmp = seg_df.groupby("count_change_bin", observed=False)["churn"].mean().reset_index()
    tmp["segment"] = seg
    temp_list.append(tmp)

count_change_effect = pd.concat(temp_list, ignore_index=True)

print("=" * 70)
print("[세그먼트 내 count_change 수준별 churn]")
print(count_change_effect)

plt.figure(figsize=(10, 6))
sns.barplot(data=count_change_effect, x="segment", y="churn", hue="count_change_bin")
plt.title("Churn Rate by Segment and Count Change Level")
plt.show()

# ---------------------------------------------------------
# 10. 핵심 인사이트 3
# - 미사용(inactive_months)의 의미를 세그먼트별로 재해석
# - 각 세그먼트 내 inactive_months 수준별 churn 비교
# ---------------------------------------------------------
temp_list = []

for seg in data["segment"].unique():
    seg_df = data[data["segment"] == seg].copy()
    
    seg_df["inactive_bin"] = pd.qcut(
        seg_df["inactive_months"],
        3,
        labels=["Low Inactive", "Mid Inactive", "High Inactive"],
        duplicates="drop"
    )
    
    tmp = seg_df.groupby("inactive_bin", observed=False)["churn"].mean().reset_index()
    tmp["segment"] = seg
    temp_list.append(tmp)

inactive_effect = pd.concat(temp_list, ignore_index=True)

print("=" * 70)
print("[세그먼트 내 inactive 수준별 churn]")
print(inactive_effect)

plt.figure(figsize=(10, 6))
sns.barplot(data=inactive_effect, x="segment", y="churn", hue="inactive_bin")
plt.title("Churn Rate by Segment and Inactive Level")
plt.show()

# ---------------------------------------------------------
# 11. 핵심 인사이트 4
# - 같은 inactive_months라도 세그먼트에 따라 churn 차이가 있는지 직접 비교
# ---------------------------------------------------------
inactive_cross = data.pivot_table(
    index="inactive_months",
    columns="segment",
    values="churn",
    aggfunc="mean"
).round(3)

print("=" * 70)
print("[inactive_months x segment churn 매트릭스]")
print(inactive_cross)

plt.figure(figsize=(10, 6))
sns.heatmap(inactive_cross, annot=True, fmt=".2f", cmap="Blues")
plt.title("Churn Rate by Inactive Months and Segment")
plt.show()

# ---------------------------------------------------------
# 12. 고객 중요도(Customer Value) 점수 정의
# - 거래금액, 거래횟수, 한도를 기반으로 중요도 산출
# ---------------------------------------------------------
value_cols = ["transaction_amount", "transaction_count", "credit_limit"]

scaled_value = scaler.fit_transform(data[value_cols])
scaled_value_df = pd.DataFrame(
    scaled_value,
    columns=[f"{c}_z2" for c in value_cols],
    index=data.index
)

data = pd.concat([data, scaled_value_df], axis=1)
data["customer_value_score"] = scaled_value_df.mean(axis=1)

data["customer_value_grade"] = pd.qcut(
    data["customer_value_score"],
    4,
    labels=["Low", "Mid-Low", "Mid-High", "High"]
)

value_summary = data.groupby("segment").agg({
    "customer_value_score": "mean",
    "transaction_amount": "mean",
    "transaction_count": "mean",
    "credit_limit": "mean",
    "churn": "mean"
}).round(3)

print("=" * 70)
print("[세그먼트별 고객 중요도]")
print(value_summary)

plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="segment", y="customer_value_score")
plt.title("Customer Value Score by Segment")
plt.show()

# ---------------------------------------------------------
# 13. 발표용 해석 테이블 자동 생성
# ---------------------------------------------------------
interpret_table = data.groupby("segment").agg({
    "transaction_count": "mean",
    "transaction_amount": "mean",
    "credit_limit": "mean",
    "utilization_ratio": "mean",
    "inactive_months": "mean",
    "count_change": "mean",
    "customer_value_score": "mean",
    "churn": "mean"
}).round(3).reset_index()

def segment_meaning(seg):
    if seg == "Dormant":
        return "저활동·저가치 고객군 / 이탈 신호에 민감"
    elif seg == "Casual":
        return "중간 활동 고객군 / 행동 변화에 따라 흔들릴 가능성"
    elif seg == "Active":
        return "고활동 고객군 / 일시적 하락과 실제 이탈 구분 필요"
    elif seg == "Premium":
        return "고한도·고가치 고객군 / 미사용이 곧 이탈은 아닐 수 있음"
    return ""

def lineup_strategy(seg):
    if seg == "Dormant":
        return "진입형·혜택형 카드"
    elif seg == "Casual":
        return "범용·포인트 카드"
    elif seg == "Active":
        return "리워드·생활밀착형 카드"
    elif seg == "Premium":
        return "프리미엄·마일리지·VIP 카드"
    return ""

interpret_table["segment_meaning"] = interpret_table["segment"].apply(segment_meaning)
interpret_table["lineup_strategy"] = interpret_table["segment"].apply(lineup_strategy)

print("=" * 70)
print("[발표 마무리용 세그먼트 해석표]")
print(interpret_table)

# ---------------------------------------------------------
# 14. 추가: 상관관계 확인
# - 발표에서 "무엇이 churn과 연관되는가" 보여주기 좋음
# ---------------------------------------------------------
corr_cols = [
    "churn",
    "transaction_count",
    "transaction_amount",
    "credit_limit",
    "utilization_ratio",
    "inactive_months",
    "count_change",
    "contact_count",
    "customer_value_score",
    "activity_score"
]

corr = data[corr_cols].corr().round(3)

print("=" * 70)
print("[상관계수]")
print(corr["churn"].sort_values(ascending=False))

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()

# ---------------------------------------------------------
# 15. 발표용 핵심 문장 자동 출력
# ---------------------------------------------------------
top_churn_segment = segment_churn.index[0]
low_churn_segment = segment_churn.index[-1]

print("=" * 70)
print("[발표용 핵심 문장]")
print(f"1. 고객 활동 수준을 기준으로 Dormant, Casual, Active, Premium의 4개 세그먼트를 도출했다.")
print(f"2. churn 비율이 가장 높은 세그먼트는 {top_churn_segment}, 가장 낮은 세그먼트는 {low_churn_segment}였다.")
print("3. 거래 변화(count_change)와 미사용(inactive_months)의 의미는 모든 고객에게 동일하지 않았으며, 세그먼트별로 다르게 나타났다.")
print("4. 특히 고활동·고가치 고객군에서는 미사용이 곧바로 이탈을 뜻하지 않을 가능성이 있으며, 저활동 고객군에서는 실제 이탈 신호일 가능성이 더 높다.")
print("5. 따라서 카드 상품 라인업은 단순한 혜택 차이가 아니라, 고객 세그먼트별 행동 특성과 가치 수준을 반영한 전략적 포트폴리오로 해석할 수 있다.")
```
